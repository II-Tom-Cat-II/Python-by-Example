```python
import random

def rand_num():
    low = int(input("Please enter a low number: "))
    high = int(input("Please enter a high number: "))
    comp_num = random.randint(low,high)
    return comp_num
    
def dialogue():
    print("I'm thinking of a number....")
    guess = int(input("What number is it? "))
    return guess
    
def check(comp_num,guess):
    try_again = True
    while try_again == True:
        if comp_num == guess:
            print("Correct!")
            try_again = False
        elif comp_num > guess:
            guess = int(input("Too low!! Try again!!"))
        else:
            guess = int(input("Too high!! Try again!!"))
            
def main():
    comp_num = rand_num()
    guess = dialogue()
    check(comp_num, guess)
    
main()
```
