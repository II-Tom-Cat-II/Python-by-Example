```python
import turtle

t1 = turtle.Turtle(shape="arrow")
t2 = turtle.Turtle(shape="arrow")
#turtle.tracer(0,0)

def pos_one():    
    turtle.ontimer(pos_one, 10)
    t1.penup()
    t1.forward(300)
    t1.left(90)
    t1.pendown()

def move_t1():
#outer ring
    turtle.ontimer(move_t1, 10)
    for i in range(0,360):
        
        t1.forward(5)
        t1.left(1)
        
def pos_two():
    turtle.ontimer(pos_two, 10)
    t2.penup()
    t2.left(90)
    t2.forward(50)
    t2.right(90)
    t2.pendown()

def move_t2():
    turtle.ontimer(move_t2, 10)
    #inner ring
    for i in range(0,330):
        t2.forward(4.5)
        t2.left(1.1)
    
def pos_three():
    turtle.ontimer(pos_three, 10)
    t1.penup()
    t1.goto((12.55,236.63))
    t1.left(150)
    t1.pendown()
    
def trig_one():
    t1.forward(415)
    t1.left(117)
    t1.forward(385)
    t1.left(118)
    t1.forward(417)
    
def pos_four():
    t2.penup()
    t2.goto(-189.89,114.99)
    t2.right(118)
    t2.pendown()
    
def trig_two():
    t2.forward(410)
    t2.right(120)
    t2.forward(399)
    t2.right(118)
    t2.forward(410)
    

turtle.ontimer(pos_one, 10)
turtle.ontimer(move_t1, 10)
turtle.ontimer(pos_two, 10)
turtle.ontimer(move_t2, 10)
    
#turtle.update()

turtle.exitonclick()
```

```python
import turtle

t1 = turtle.Turtle(shape="turtle")
t2 = turtle.Turtle(shape="turtle")

t1.setheading(45)
t2.setheading(-135)

def move_t1():
    t1.forward(1)
    turtle.ontimer(move_t1, 10)

def move_t2():
    t2.forward(1)
    turtle.ontimer(move_t2, 10)

turtle.ontimer(move_t1, 10)
turtle.ontimer(move_t2, 10)

turtle.exitonclick()
```
